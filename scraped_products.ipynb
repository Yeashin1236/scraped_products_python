```python
for current_page in range(1, 49):
    print(f"Currently scraping page: {current_page}")
    url = url.format(current_page)
    
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to fetch page {current_page}")
        continue

    soup = BeautifulSoup(response.text, 'html.parser')

    # Check if page not found
    if soup.title and 'Page not found' in soup.title.string:
        print(f"Page {current_page} not found, stopping.")
        break

    product_items = soup.find_all("li", class_="product")

    for item in product_items:
        try:
            name = item.find("h2", class_="woocommerce-loop-product__title").text.strip()
            price = item.find("span", class_="woocommerce-Price-amount amount").text.strip().replace("£", "").replace(".00", "")
            url = item.find("a")["href"]

            data.append({
                "Name": name,
                "Price": price,
                "URL": url
            })
        except Exception as e:
            print(f"Error parsing item: {e}")
            continue

    time.sleep(1)  # Be polite to the server
```

```python
# Create DataFrame
products_df = pd.DataFrame(data)
print(products_df)
```

```python
# Save to CSV and excel 
products_df.to_csv(r"F:\Python project\Scrape Me\scraped_products.csv", index=False)
products_df.to_excel(r"F:\Python project\Scrape Me\scraped_products.xlsx", index=False)
```
